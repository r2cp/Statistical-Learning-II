## Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Carga de datos

```python
# Cargamos los datos desde el archivo csv
datos = pd.read_csv("data.csv")
print(datos.head())

# Mostramos como están distribuidos los datos en una gráfica
plt.figure(figsize=(14,8))
plt.scatter(datos.x1[:], datos.x2[:], alpha=0.8);
plt.xlabel("X1");
plt.ylabel("X2");
```

## Algoritmo de entrenamiento del perceptrón

```python
# Agregamos el vector de 1's para el término independiente
datos['bias'] = np.ones(len(datos))
```

```python
# Función para entrenar el perceptrón
def entrenarPerceptron(x, y, lr, epochs, z=0):
    #Inicializamos el vector de pesos sinápticos
    w = np.random.standard_normal(size=x.shape[1])
    # Iteraciones
    n = 0        
    # Mientras no sobrepasemos un límite de iteraciones
    while n < epochs:          
        for i in range(len(x)):    
            # Producto punto entre x y el vector de pesos sinápticos
            s = np.dot(x[i], w)
            yhat = 1. if s >= z else 0.
            for j in range(len(w)):             
                w[j] = w[j] + lr*(y[i]-yhat)*x[i][j]              
        n += 1
    # Devolvemos el vector de ponderaciones
    return w
```

```python
# Seleccionamos los datos para X
x = np.array(datos[['bias', 'x1', 'x2']])
# Seleccionamos los datos para Y
y = datos['label']

# Entrenamiento del perceptrón
w_ = entrenarPerceptron(x, y, lr=0.1, epochs=1000)
```

```python
# Mostramos los resultados de correr nuestros datos
w_
```

```python
# Definimos una función con los datos obtenidos por el perceptron
def decisionBoundary(x1):
    return (-(pesos[0] + pesos[1]*x1)/ pesos[2])

x1 = np.linspace(0,20)

# Graficamos los datos y el decision boundary
plt.figure(figsize=(14,8))
plt.scatter(datos.x1[:], datos.x2[:], alpha=0.8, label="Datos de entrenamiento")
plt.plot(x1, np.array([decisionBoundary(x) for x in x1]), 'r--', label="Decision boundary")
plt.xlabel("X1");
plt.ylabel("X2");
plt.legend()
plt.show()
```
